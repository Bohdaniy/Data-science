```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. NumPy
# Створення даних за допомогою arange та random
data_scalar = np.random.randint(1, 10)  # Скаляр
data_vector = np.arange(0, 10)  # Вектор
data_matrix = np.random.rand(3, 3)  # Матриця

# CRUD операції
# Створення
new_data = np.random.rand(3, 3)
# Читання
read_scalar = data_scalar
read_vector = data_vector[0:5]
read_matrix = data_matrix[:, 1]
# Оновлення
data_vector[0] = 100
# Видалення
data_matrix = np.delete(data_matrix, 0, axis=0)

print(f"Скаляр: {read_scalar}")
print(f"Вектор (частина): {read_vector}")
print(f"Матриця (стовпчик 1): {read_matrix}")

# 2. Pandas
# Створення DataFrame різними способами
df1 = pd.DataFrame(data_vector, columns=["Values"])
df2 = pd.DataFrame({
    'A': np.random.randint(1, 10, size=10),
    'B': np.random.random(size=10)
})

# Маніпуляції з даними
print(df1.head())  # Перші 5 рядків
print(df2.describe())  # Опис статистичних даних
print(df2.iloc[0:3])  # Вибір даних за індексом
print(df2.loc[df2['A'] > 5])  # Вибір рядків, де значення A більше 5

# 3. Matplotlib та обчислення похибок
# Створення лінійної функції
X = np.linspace(0, 10, 100)
Y = 2 * X + 1  # Лінійна функція
# Додавання похибок
Y_noise = Y + np.random.normal(0, 1, size=Y.shape)

# Побудова графіку
plt.plot(X, Y, label='Лінійна функція')
plt.scatter(X, Y_noise, label='Дані з похибками', color='red', s=10)
plt.legend()
plt.show()

# Обчислення MAE та MSE
mAE = mean_absolute_error(Y, Y_noise)
mSE = mean_squared_error(Y, Y_noise)

# Запис результатів у CSV
results = pd.DataFrame({
    'X': X,
    'Y': Y,
    'Y_hat': Y_noise,
    'MAE': [mAE] * len(X),
    'MSE': [mSE] * len(X)
})

results.to_csv('results.csv', index=False)

print(f"Mean Absolute Error: {mAE}")
print(f"Mean Squared Error: {mSE}")
```
